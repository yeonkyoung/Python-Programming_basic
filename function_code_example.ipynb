# filter, map, reduce
- filter : 특정 조건을 만족하는 요소만 남기고 필터링
- map : 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 반환 (함수를 mapping해서 반환)
- reduce : 차례대로 앞 2개의 원소를 가지고 연산. 연산의 결과가 또 다음 연산의 입력으로 진행. 따라서 마지막까지 진행되면 최종 출력은 한개의 값만 남게 됨

```python
nums=[1,2,3,6,8,9]
#짝수만 
list(filter(lambda a:a%2==0 ,nums))
```

```python
#제곱
list(map(lambda a:a**2,nums))
```

```python
import functools

a=[1,3,5,8]

#리스트내의 모든 숫자의 합
functools.reduce(lambda x,y:x+y,a)
```

# 클래스와 인스턴스
- class : 실세계의 것을 모델링하여 속성, 동작을 갖는 데이터 '타입'. 다루고자하는 데이터와 함수를 하나로 캡슐화 하여 클래스로 표현. 모델링에서 중요시 하는 속성에 따라 클래스의 속성과 행동이 각각 달라짐

- init(self) : 생성자,클래스 인스턴스가 생성될 때 호출됨. self인자는 항상첫번째에 오며 자기자신을 가르킴.

```python
class person : #class 선언
    def __init__(self):
        print(self , 'is generated')
        self.name= 'kate' #속성1
        self.age= 10 #속성2
        
p1=person()
p2=person() # 두가지 객체 생성됨

p1.name= 'aron' #속성을 외부에서 바꾸기
p2.age= 20

print(p1.name,p1.age)
```

```python
#속성을 클래스에서 입력하여 바꾸기
class person : #class 선언
    def __init__(self,n,a):
        print(self , 'is generated')
        self.name= n
        self.age= a

p1 = person('Bob',30)
p2 = person('Kate',20)

print(p1.name,p1.age)
print(p2.name,p2.age)
        
```

# method,static method 정의 사용하기
- method 정의 : 멤버함수라고도 하며, 해당 클래스의 object에서만 호출 가능
- 메쏘드는 객체 레벨에서 호출되며, 해당 객체의 속성에 대한 연산을 행함
- {obj}.{method}() 형태로 호출됨




```python
#1.숫자를 하나 증가
#2.숫자를 0으로 초기화
class Counter:
    def __init__(self):
        self.num=0
    def increment(self):
        self.num += 1
    def reset(self):
        self.num=0
    def print_current_value(self):
        print('현재값은:',self.num)
        
        
c1=Counter() #호출되는 레벨이 객체임
c1.print_current_value()
c1.increment()
c1.increment()
c1.print_current_value()
c1.reset()
c1.print_current_value()

c2=Counter() #호출되는 레벨이 객체임
c2.increment()
c2.print_current_value()
```

# method type
- instantce method : 객체로 호출. 매쏘드는 객체레벨로 호출되기 때문에, 해당 메쏘드를 호출한 객체에만 영향을 미침
- class method(static method) : class로 호출. 클래스 레벨로 호출되기 떄문에, 클래스 멤버 변수만 변경 가능

```python
class Math:
    @staticmethod
    def add(a,b):
        return a+b
    @staticmethod
    def multiply(a,b):
        return a*b


Math.add(10,20)
Math.multiply(10,20)
```

# method override 
- 부모 클래스의 method를 재정의(override). 하위클래스(자식클래스)의 인스턴스로 호출시, 재정의된 메소드가 호출됨

```python
class Person :
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def eat(self,food) :
        print('{}은 {}를 먹습니다.'.format(self.name,food))
        
    def sleep(self,minute):
        print('{}은 {}분을 잡니다.'.format(self.name,minute))
        
    def work(self,minute):
        print('{}은 {}분을 일합니다.'.format(self.name,minute))
    
class Student(Person) : #person클래스 상속
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def work(self,minute): #work속성 재정의
        print('{}은 {}분을 공부합니다.'.format(self.name,minute))
    
class Employee:
    def __init__(self,name,age):#person클래스 상속
        self.name=name
        self.age=age
        
    def work(self,minute): #work속성 재정의
        print('{}은 {}분을 업무를합니다.'.format(self.name,minute))


bob=Student('bob',25)
bob.eat('BBQ')
bob.sleep(30)
bob.work(60)
```

# super
- 하위클래스(자식 클래스)에서 부모 클래스의 method를 호출할 때 사용

```python
class Person :
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def eat(self,food) :
        print('{}은 {}를 먹습니다.'.format(self.name,food))
        
    def sleep(self,minute):
        print('{}은 {}분을 잡니다.'.format(self.name,minute))
        
    def work(self,minute):
        print('{}은 {}분동안 준비를 합니다.'.format(self.name,minute))
    
class Student(Person) : #person클래스 상속
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def work(self,minute): 
        super().work(minute) #부모 속성 불러오기
        print('{}은 {}분을 공부합니다.'.format(self.name,minute)) #속성 재정의
        # 부모속성과 재정의한 속성이 동시 출력된다
    
class Employee:
    def __init__(self,name,age):#person클래스 상속
        self.name=name
        self.age=age
        
    def work(self,minute): #work속성 재정의
        super().work(minute)
        print('{}은 {}분을 업무를합니다.'.format(self.name,minute))


bob=Student('bob',25)
bob.eat('BBQ')
bob.sleep(30)
bob.work(60)
```

# 클래스 연산자 재정의 이해 및 사용(오버라이딩 함수 이용)
- https://docs.python.org/3/reference/datamodel.html#object.__add__



```python
#2차원 좌표평면 각점(x,y)
#연산(두점의 덧셈,뺄셈,곱셈)
#그점의 길이(0,0)으로 부터 거리
#x,y값 가져오기 (인덱스로)
#출력하기

class Point :
    def __init__(self,x,y):
        self.x=x
        self.y=y
    
    def __add__(self,pt):
        new_x=self.x + pt.x
        new_y=self.y + pt.y
        return Point(new_x,new_y)
    
    def __sub__(self,pt):
        new_x=self.x - pt.x
        new_y=self.y - pt.y
        return Point(new_x,new_y)
    
    def __mul__(self,factor): #곱셈
        return Point(self.x * factor ,self.y * factor)
    
    def __len__(self): #길이 구하기
        return self.x ** 2 + self.y ** 2
    
    def __getitem__(self,index):
        if index == 0 :
            return self.x
        elif index == 1:
            return self.y
        else:
            return -1
        
    def __str__(self): # 속성 print대신 __str__ 사용
        return '({},{})'.format(self.x,self.y)
    
    
p1=Point(3,4)
p2=Point(5,6)
p3=p1+p2
p4=p1-p2
p5=p1 * 3

print(p1)
print(p2)
print(p3)
print(p4)
print(p5)
print(len(p1))
print(p1[0])
print(p1[1])

```
